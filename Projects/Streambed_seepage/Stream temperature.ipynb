```python
doc_dir = os.getcwd()
while os.path.basename(doc_dir) != 'Documents':
    doc_dir = os.path.dirname(doc_dir)
# dir of all gwfm data
gwfm_dir = os.path.dirname(doc_dir)+'/Box/research_cosumnes/GWFlowModel'
# dir of stream level data for seepage study
proj_dir = gwfm_dir + '/Stream_seepage/'
lvl_dir = proj_dir+'Stream_level_data/'
```

```python
os.listdir(gwfm_dir+'/Mapping')
rm = pd.read_csv(gwfm_dir+'/Mapping/allsensor_latlong.csv')
rm = rm[rm.Group=='River']
rm_gdf = gpd.GeoDataFrame(rm, geometry = gpd.points_from_xy(rm.Longitude, rm.Latitude))
rm_gdf.crs='epsg:4326'
rm_gdf = rm_gdf.to_crs('epsg:32610')
rm_gdf = rm_gdf[rm_gdf.Site_id.str.contains('RM')]

rm_gdf.plot('Site_id',legend=True)
```

```python
#Pw (water vapor pressure) = RH (relative humidity)× Pws (saturation water vapor pressure)⁄ 100

# load air temperature data
air = pd.read_csv(proj_dir+'cimis_daily_all_sensors_2019_2022.csv', parse_dates=['Date'])
air.columns
cols = ['Date','Precip (mm)','Net Rad (W/sq.m)','Max Air Temp (C)','Min Air Temp (C)','Avg Air Temp (C)']
air = air[cols]
air = air.set_index('Date')
```

```python
os.listdir(lvl_dir)
strm_dat = pd.read_csv(lvl_dir+'Stream_level_m_temp_C_long.csv', parse_dates = ['dt'])
strm_dat.columns
```

```python
lvl = strm_dat[strm_dat.param=='LEVEL'].pivot_table(values='value', index='dt', columns='site_nam')
# create equal adjustment among sensors
lvl.loc[:,lvl.min()>9.5] -= 9.5
```

```python
lvl.plot()
```

```python
tmp = strm_dat[strm_dat.param=='TEMPERATURE'].pivot_table(values='value', index='dt', columns='site_nam')
tmp.plot()

air.plot(y='Avg Air Temp (C)')
```

```python
(tmp.resample('D').max()-tmp.resample('D').min()).mean(axis=1).plot()
underwater = (tmp.resample('D').max()-tmp.resample('D').min()).mean(axis=1) 
underwater = underwater.resample('15min').pad()
underwater = underwater[underwater<5]
underwater = underwater[underwater.index.isin(lvl.index)]
```

```python
# create consistent color scheme
rm_gdf['num'] = rm_gdf.Site_id.str.extract(pat = r'(\d+)').astype(int)


from matplotlib import colors
from matplotlib import cm 
# do the colour mapping
c_norm  = colors.Normalize(vmin=rm_gdf.num.min(), vmax=rm_gdf.num.max())
scalar_map = cm.ScalarMappable(norm=c_norm, cmap=plt.get_cmap('viridis'))    

rm_gdf['colors'] = [scalar_map.to_rgba(val) for val in rm_gdf.num.values]
rm_colors = rm_gdf.set_index('Site_id')['colors']
```
