```python
import geopandas as gpd
import rasterio
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import shapely
import os
from flopy.utils import Raster
```

```python
mb_name = os.path.dirname(os.getcwd())+"\GWModelDomain_UTM10N\GWModelDomain_Rec_UTM10N.dbf"
# Rivers and creeks in the larger area encompassing Cosumnes River in both South American and Cosumnes Subbasins
rivers = gpd.read_file("Sac_valley_rivers/Sac_valley_rivers.shp")
mb = gpd.read_file(mb_name)
rivers = rivers.to_crs('EPSG:32610')
rivers_clip = gpd.clip(rivers, mb)
rivers_clip.plot()
# rivers_clip.GNIS_Name.unique()
```

```python
print(rivers_clip.columns)
# Split into individual streams/creeks
cr_ind = rivers_clip[rivers_clip.GNIS_Name == 'Cosumnes River']
dc_ind = rivers_clip[rivers_clip.GNIS_Name == 'Deer Creek']
cc_ind = rivers_clip[rivers_clip.GNIS_Name ==  'Coyote Creek']
# Pull out data for each river/creek
cr = rivers_clip.loc[cr_ind.index,]
dc = rivers_clip.loc[dc_ind.index,]
cc = rivers_clip.loc[cc_ind.index,]
```

```python
import fiona
from shapely.geometry import shape, mapping
from shapely.ops import linemerge

cr.geometry.values.crs = "epsg:32610"
geom = linemerge(cr.geometry.values)
# how often to interpolate a point
dline = 10
# # length of the LineString
length = int(geom.length)
point = np.zeros((int(length/dline)+1,3))
for i, distance in enumerate(range(0, int(length), dline)):
         point[i,:] = geom.interpolate(distance).coords[:][0]
point = point[:,[0,1]]
plt.plot(point[:,0],point[:,1])
```

```python
spath = "C://Users/ajcalder/Box/Research_Calderwood/dem"
raster_name = spath+'/USGS_ten_meter_dem/regional_10m.tif'

# dem10 = rasterio.open(raster_name)

pnts = pd.DataFrame()
with rasterio.open(raster_name) as src:
    pnts['z10'] = [sample[0] for sample in src.sample(point)]
pnts
pnts['easting'] = point[:,0]
pnts['northing'] = point[:,1]
pnts = pnts[pnts.z10 > -1E4]
pnts_original = pnts.copy()
```

```python
fig, ax = plt.subplots(figsize = (10,10))
# pnts.z10.plot()
pnts = pnts_original.copy()
pnts['slope'] = 0.002
for i in np.arange(1,len(pnts)):
    if pnts.z10.values[i] > pnts.z10.values[i-1]:
        # if strtop is greater than previous strtop use previous elevation minus the average slope
        slope = ((np.max(pnts.z10) - np.min(pnts.z10))/geom.length)*dline
        if pnts.index[i] < 800:
            slope = 0.002
        elif pnts.index[i] < 2700:
            slope = 0.0005
        elif pnts.index[i] < 3200:
            slope = 0.001
        else:
            slope = 0.0001
        pnts.z10.values[i] = pnts.z10.values[i-1] - slope*dline
        pnts.slope.values[i] = slope

        
pnts_original.z10.plot()
pnts.z10.plot()
```

## Convert xy of terrain.cos.tif from feet to meters
This portion of the code only needs to be run once and has already been completed

```python
import numpy as np
import rasterio
from rasterio.warp import calculate_default_transform, reproject, Resampling
raster_name = "C:\\Users\\ajcalder\\Box\\Research_Calderwood\\Cos\\Terrain\\Terrain.Cos.tif"
dst_crs = 'EPSG:26910'

with rasterio.open(raster_name) as src:
    transform, width, height = calculate_default_transform(
        src.crs, dst_crs, src.width, src.height, *src.bounds)
    kwargs = src.meta.copy()
    kwargs.update({
        'crs': dst_crs,
        'transform': transform,
        'width': width,
        'height': height
    })

    with rasterio.open('terrain.cos.meters.tif', 'w', **kwargs) as dst:
        for i in range(1, src.count + 1):
            reproject(
                source=rasterio.band(src, i),
                destination=rasterio.band(dst, i),
                src_transform=src.transform,
                src_crs=src.crs,
                # Important Part Here
                # I adjust the transform so that it is in units of meters instead of feet
                # The scale is already transformed from 3.125 units per cell to ~1 unit per cell
                dst_transform=src.transform.translation(src.transform[2]/3.28084, src.transform[5]/3.28084),
                dst_crs=dst_crs,
                resampling=Resampling.nearest)
```
