```python
# updated function because unary_union is slow for large spatial data
from shapely.geometry import  box
def gdf_bnds(gdf, ax=None, buf=None):
    gdf_bnd = gpd.GeoDataFrame([0], geometry=[box(*gdf.total_bounds)], crs = gdf.crs)
    if buf != None:
        gdf_bnd.geometry = gdf_bnd.buffer(buf)
    if ax !=None:
        xmin, ymin, xmax, ymax = gdf_bnd.total_bounds
        ax.set_xlim(xmin,xmax)
        ax.set_ylim(ymin, ymax)
    return(gdf_bnd)
```

```python
# read the shapefiles (** indicates 0-n sub directories)
fn = glob.glob(join(gis_dir,'flight_lines','**','*.shp'), recursive=True)
```

```python
flight_df = pd.DataFrame(fn, columns=['path'])
flight_df['fn'] = flight_df.path.apply(basename)
flight_df['survey'] = flight_df.fn.str.extract(r'(\d)').astype(int)
```

```python
# select the region to analyze
survey = 1
```

```python
lines = gpd.read_file(flight_df[flight_df.survey==survey].path.values[0])
```

```python
## identify which rivers can be analyzed
riv = rivers.to_crs(lines.crs)
# crops to boundary extent
riv = gpd.overlay(riv, bnds)
riv_lines = gpd.sjoin(riv, lines,predicate='intersects', how='inner', rsuffix='lines')
# only count lines once for the number of times that they cross a river (avoid double counts for sinuosity)
riv_lines = riv_lines.drop_duplicates(['GNIS_Name', 'index_lines'])
# identify how many lines that a river has overlapping it
riv_lines_count = riv_lines.groupby('GNIS_Name').count()['index_lines']
print(riv_lines_count)
# start with 10 as an arbitraty number of lines needed before anlysis is done
# necessary to provide significant data (avoid local bias) and show trends
riv_keep = riv_lines_count[riv_lines_count>10].index
# subset rivers to only those with significant line crossing
riv = riv[riv.GNIS_Name.isin(riv_keep)]
```

```python
r_name = riv_keep[0]
# if there are multiple rivers identified then need to iterate over rivers to be consistent
riv_plt = riv[riv.GNIS_Name==r_name]
riv_plt = gpd.GeoDataFrame([0], geometry = [riv_plt.unary_union], crs=  riv_plt.crs)
```

```python
# create buffered version of river for setback analysis
riv_buf = riv_plt.copy()

riv_buf.geometry = riv_buf.buffer(3200)
```

```python
fig,ax = plt.subplots(figsize=(6.5,8))
# bnds = gdf_bnds(lines, ax=ax, buf=1E3)
bnds = gdf_bnds(riv, ax=ax, buf=1E3)
riv_plt.plot(ax=ax, linewidth=.5)
riv_buf.plot(ax=ax, color='none',linewidth=.5)

lines.plot(ax=ax, linewidth=0.5, alpha=0.5, color='brown')

plt_cln(ax=ax)
# ctx.add_basemap(ax=ax, source = ctx.providers.Esri.WorldImagery, attribution=False, attribution_size=6,
#                 crs = 'epsg:3310', alpha=0.8)
```
